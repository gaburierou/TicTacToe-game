```python
from tkinter import *
from tkinter import messagebox
import random

version = "0.5"

class Hash(Frame):
    def __init__(self, master):
        Frame.__init__(self, master)

        self.areas = None
        
        self.pinned = None
        
        self.create_widgets()
        
        self.start_variable()
        
        
    def start_variable(self):
        self.areas = ['one', 'two', 'three',
                       'four', 'five', 'six',
                       'seven', 'eight', 'nine']
        
        self.pinned = []
        
        self.one_show["text"] = "  "
        self.two_show["text"] = "  "
        self.three_show["text"] = "  "
        self.four_show["text"] = "  "
        self.five_show["text"] = "  "
        self.six_show["text"] = "  "
        self.seven_show["text"] = "  "
        self.eight_show["text"] = "  "
        self.nine_show["text"] = "  "

    def get_number(self, n):
        win = False
        
        lose = False
        
        draw = False
 
        x = n               
                    
        if x == self.areas[self.get_index(self.areas, x)]:


            self.pinned.append(self.areas[self.get_index(self.areas, x)])
            
            if x == "one":
                self.one_show["text"] = "X"
            
            if x == "two":
                self.two_show["text"] = "X"
            
            if x == "three":
                self.three_show["text"] = "X"
            
            if x == "four":
                self.four_show["text"] = "X"
            
            if x == "five":
                self.five_show["text"] = "X"
            
            if x == "six":
                self.six_show["text"] = "X"
            
            if x == "seven":
                self.seven_show["text"] = "X"
            
            if x == "eight":
                self.eight_show["text"] = "X"
            
            if x == "nine":
                self.nine_show["text"] = "X"
            
            self.areas[self.get_index(self.areas, x)] = 0                                
                
            self.areas[self.random(self.areas, x)] = 1

                
        else:
            messagebox.showinfo("Info","Invalid value!")
            
            
        #player verification:

        #horizontal    
        if self.areas[0] == 0 and self.areas[1] == 0 and self.areas[2] == 0:
            win = True
        if self.areas[3] == 0 and self.areas[4] == 0 and self.areas[5] == 0:
            win = True
        if self.areas[6] == 0 and self.areas[7] == 0 and self.areas[8] == 0:
            win = True

        #vertical
        if self.areas[0] == 0 and self.areas[3] == 0 and self.areas[6] == 0:
            win = True
        if self.areas[1] == 0 and self.areas[4] == 0 and self.areas[7] == 0:
            win = True
        if self.areas[2] == 0 and self.areas[5] == 0 and self.areas[8] == 0:
            win = True

        #diagonal
        if self.areas[0] == 0 and self.areas[4] == 0 and self.areas[8] == 0:
            win = True
        if self.areas[2] == 0 and self.areas[4] == 0 and self.areas[6] == 0:
            win = True


        #engine verification:

        #horizontal    
        if self.areas[0] == 1 and self.areas[1] == 1 and self.areas[2] == 1:
            lose = True
        if self.areas[3] == 1 and self.areas[4] == 1 and self.areas[5] == 1:
            lose = True
        if self.areas[6] == 1 and self.areas[7] == 1 and self.areas[8] == 1:
            lose = True

        #vertical
        if self.areas[0] == 1 and self.areas[3] == 1 and self.areas[6] == 1:
            lose = True
        if self.areas[1] == 1 and self.areas[4] == 1 and self.areas[7] == 1:
            lose = True
        if self.areas[2] == 1 and self.areas[5] == 1 and self.areas[8] == 1:
            lose = True

        #diagonal
        if self.areas[0] == 1 and self.areas[4] == 1 and self.areas[8] == 1:
            lose = True
        if self.areas[2] == 1 and self.areas[4] == 1 and self.areas[6] == 1:
            lose = True
        
        #draw:       
        if lose == False and win == False and len(self.pinned) == 9:
            draw = True
                           
        #notifications:
        if win:          
            messagebox.showinfo("Congrats","YOU WIN!")
            self.start_variable()
        if lose:
            messagebox.showinfo("Info","You lose :(")
            self.start_variable()
        if draw:
            messagebox.showinfo("Info","DRAW")
            self.start_variable()
        
    def get_index(self, itens, n):     
        for i in itens:
            if n not in itens:
                return -1
            else:
                return itens.index(n)
            
    def random(self, itens, x):
        while True:
            if len(self.pinned) >= 8:
                no = itens.index(1)
                return no
                break
            n = random.choice(itens)
            if n != x and n != 0 and n != 1:                    
                self.pinned.append(n)
                
                if n == "one":
                    self.one_show["text"] = "O"
            
                if n == "two":
                    self.two_show["text"] = "O"
            
                if n == "three":
                    self.three_show["text"] = "O"

                if n == "four":
                    self.four_show["text"] = "O"

                if n == "five":
                    self.five_show["text"] = "O"

                if n == "six":
                    self.six_show["text"] = "O"

                if n == "seven":
                    self.seven_show["text"] = "O"

                if n == "eight":
                    self.eight_show["text"] = "O"

                if n == "nine":
                    self.nine_show["text"] = "O"
                
                return itens.index(n)
                break
                
        
        
    def create_widgets(self):
        #show game
        self.frame1 = Frame(self, relief = SUNKEN, borderwidth=10)
        self.frame1.grid(row=0)
        
        self.frame2 = Frame(self, relief = SUNKEN, borderwidth=10)
        self.frame2.grid(row=0, column=1)
        
        self.frame3 = Frame(self, relief = SUNKEN, borderwidth=10)
        self.frame3.grid(row=0, column=2)
        
        
        
        self.frame4 = Frame(self, relief = SUNKEN, borderwidth=10)
        self.frame4.grid(row=1, column=0)
        
        self.frame5 = Frame(self, relief = SUNKEN, borderwidth=10)
        self.frame5.grid(row=1, column=1)
        
        self.frame6 = Frame(self, relief = SUNKEN, borderwidth=10)
        self.frame6.grid(row=1, column=2)
        
        
        
        self.frame7 = Frame(self, relief = SUNKEN, borderwidth=10)
        self.frame7.grid(row=2, column=0)
        
        self.frame8 = Frame(self, relief = SUNKEN, borderwidth=10)
        self.frame8.grid(row=2, column=1)
        
        self.frame9 = Frame(self, relief = SUNKEN, borderwidth=10)
        self.frame9.grid(row=2, column=2)
        
        
        
        self.one_show = Label(self.frame1, font=("Verdana", 12), text="  ", borderwidth=10,padx=20,pady=10)
        self.one_show.grid(row=0, column=0, sticky="NWNESWSE")
        
        self.two_show = Label(self.frame2, font=("Verdana", 12),text="  ", borderwidth=10,padx=20,pady=10)
        self.two_show.grid(row=0, column=1, sticky="NWNESWSE")
        
        self.three_show = Label(self.frame3, font=("Verdana", 12), text="  ", borderwidth=10,padx=20,pady=10)
        self.three_show.grid(row=0, column=2, sticky="NWNESWSE")
        
        
        
        self.four_show = Label(self.frame4, font=("Verdana", 12), text="  ", borderwidth=10,padx=20,pady=10)
        self.four_show.grid(row=1, column=0, sticky="NWNESWSE")
        
        self.five_show = Label(self.frame5, font=("Verdana", 12), text="  ", borderwidth=10,padx=20,pady=10)
        self.five_show.grid(row=1, column=1, sticky="NWNESWSE")
        
        self.six_show = Label(self.frame6, font=("Verdana", 12), text="  ", borderwidth=10,padx=20,pady=10)
        self.six_show.grid(row=1, column=2, sticky="NWNESWSE")
        
        
        
        self.seven_show = Label(self.frame7, font=("Verdana", 12), text="  ", borderwidth=10,padx=20,pady=10)
        self.seven_show.grid(row=2, column=0, sticky="NWNESWSE")
        
        self.eight_show = Label(self.frame8, font=("Verdana", 12), text="  ", borderwidth=10,padx=20,pady=10)
        self.eight_show.grid(row=2, column=1, sticky="NWNESWSE")
        
        self.nine_show = Label(self.frame9, font=("Verdana", 12), text="  ", borderwidth=10,padx=20,pady=10)
        self.nine_show.grid(row=2, column=2, sticky="NWNESWSE")
        
        #show buttons
        
        self.button1 = Button(self.frame1, font=("System", 6), text="x", padx=10, pady=0, command=lambda: self.get_number("one"))
        self.button1.grid(row=1, column=0)
        
        self.button2 = Button(self.frame2, font=("System", 6), text="x", padx=10, pady=0, command=lambda: self.get_number("two"))
        self.button2.grid(row=1, column=1)
        
        self.button3 = Button(self.frame3, font=("System", 6), text="x", padx=10, pady=0, command=lambda: self.get_number("three"))
        self.button3.grid(row=1, column=2)
        
        self.button4 = Button(self.frame4, font=("System", 6), text="x", padx=10, pady=0, command=lambda: self.get_number("four"))
        self.button4.grid(row=4, column=0)
        
        self.button5 = Button(self.frame5, font=("System", 6), text="x", padx=10, pady=0, command=lambda: self.get_number("five"))
        self.button5.grid(row=4, column=1)
        
        self.button6 = Button(self.frame6, font=("System", 6), text="x", padx=10, pady=0, command=lambda: self.get_number("six"))
        self.button6.grid(row=4, column=2)
        
        self.button7 = Button(self.frame7, font=("System", 6), text="x", padx=10, pady=0, command=lambda: self.get_number("seven"))
        self.button7.grid(row=5, column=0)
        
        self.button8 = Button(self.frame8, font=("System", 6), text="x", padx=10, pady=0, command=lambda: self.get_number("eight"))
        self.button8.grid(row=5, column=1)
        
        self.button9 = Button(self.frame9, font=("System", 6), text="x", padx=10, pady=0, command=lambda: self.get_number("nine"))
        self.button9.grid(row=5, column=2)
        
        #restart
        self.frame10 = Frame(self, relief = SUNKEN, borderwidth=2)
        self.frame10.grid(row=6, column=1)
        
        self.button10 = Button(self.frame10,font=("System", 12), text="reset", padx=10, pady=0, command=lambda: self.start_variable())
        self.button10.grid(row=6, column=1)
        
        
        
        
hash = Tk()
hash.title(f"Hash Game {version}")
hash.resizable(0, 0)
app = Hash(hash).grid()
hash.mainloop()
```
